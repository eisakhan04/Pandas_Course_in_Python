**Pivot_Table Function**

```python
import pandas as pd
a=pd.read_csv('Data.csv.csv')
a
```

```python
b = a.pivot_table(index = 'Country')
b
```

```python
b = a.pivot_table(index = 'Country',columns = 'Age')
b
```

```python
b = a.pivot_table(index = 'Country',columns = 'Age',aggfunc= 'count')
b
```

```python
b = a.pivot_table(index = 'Country',columns = 'Age',aggfunc= 'sum')
b
```

```python
b = a.pivot_table(index = 'Country',columns = 'Age',fill_value=0)
b
```
