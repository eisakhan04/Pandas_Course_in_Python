# **DataFrame.aggregate**

```python
import pandas as pd
import numpy as np
data1 = pd.DataFrame({'a': [1, 2, 3, 4, 5], 'b': [6, 7, 8, 9, 10]})
print("Data1:")
print(data1)
print('Aggregate')
data2 = data1.agg('sum')
print(data2)
```

```python
import pandas as pd
import numpy as np
data1 = pd.DataFrame({'a': [1, 2, 3, 4, 5], 'b': [6, 7, 8, 9, 10]})
print("Data1:")
print(data1)
print('Aggregate')
data2 = data1.aggregate(['mean', 'median','max','min','sum','std','var','corrcoef','ptp'])
print(data2)
```
